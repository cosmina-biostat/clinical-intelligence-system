# Notebook 04 — Train all classifiers
Trains LogReg, RF, XGBoost, LightGBM on all 3 datasets. Saves .pkl files to models/saved/.

```python
import sys; sys.path.insert(0, '..')
from src.data.loader import load_dataset
from src.data.preprocessor import split_and_balance
from src.models.train import train_all
from src.models.evaluate import comparison_table, plot_roc_curves
import pandas as pd
```

```python
all_results = {}

for dataset_name, target_col in [("cardio","target"),("ms","target"),("melanoma","target")]:
    df = load_dataset(dataset_name)
    X  = df.drop(columns=[target_col])
    y  = df[target_col]
    X_train, X_test, y_train, y_test = split_and_balance(X, y, dataset_name)
    results = train_all(X_train, y_train, X_test, y_test, dataset_name)
    all_results[dataset_name] = results
    print(f'{dataset_name} done: {results}')
```

```python
table = comparison_table(all_results)
table
```
